```python
import pandas as pd
import os.path
import numpy as np
import numba as nb
import gc
```

```python
file_path = '../Fixed Census Ontario 2021.csv'
original_path = '../Census Data Ontario 2021.csv'
```

```python
# Helper function to save the data from RAM to Disk
def append_to_file(file_path, to_save_df):
    header = True
    # if file exists then don't append the header again
    if os.path.isfile(file_path):
        header = False
        
    to_save_df.to_csv(file_path, mode='a', header=header, index=False)
    
chunk_size = 256_961
chunk_iter = 896
```

```python
use_cols = ["ALT_GEO_CODE", "CHARACTERISTIC_NAME", "C1_COUNT_TOTAL", "GEO_NAME"]
use_cols_2016 = [
    "ALT_GEO_CODE",
    "DIM: Profile of Dissemination Areas (2247)",
    "Dim: Sex (3): Member ID: [1]: Total - Sex",
    "GEO_NAME",
]
df = pd.read_csv(
    original_path, encoding="latin1", nrows=3000, low_memory=True, usecols=use_cols
)

# Gets the labels for the new data frame
labels = tuple(df.loc[0:2631, "CHARACTERISTIC_NAME"].tolist())

del df

print("Deleted Df")
count = 0


@nb.jit
def process_row(df):
    df_t = df.set_index("CHARACTERISTIC_NAME").T.reset_index()
    df_t["ALT_GEO_CODE"] = df_t.iloc[1,2]
    df_t["GEO_NAME"] = df_t.iloc[0,3]
    df_t.drop([0,1], axis=0, inplace=True)
    df_t.drop("index", axis=1, inplace=True)
    return df_t


for df in pd.read_csv(
    original_path,
    encoding="latin1",
    chunksize=2631,
    low_memory=True,
    usecols=use_cols,
):
    # df.rename(
    #     columns={
    #         "DIM: Profile of Dissemination Areas (2247)": "CHARACTERISTIC_NAME",
    #         "Dim: Sex (3): Member ID: [1]: Total - Sex": "C1_COUNT_TOTAL",
    #     },
    #     inplace=True,
    # )
    count += 1
    print(f"Loaded Chunk: {count}")
    new_df = process_row(df)
    append_to_file(file_path, new_df)
    del new_df, df
    gc.collect()
```

```python

df = pd.read_csv(
    original_path, encoding="latin1", low_memory=True, usecols=use_cols
)
unique_codes = df['ALT_GEO_CODE'].unique()
append_df = pd.DataFrame()
for code in unique_codes:
    print(f"Working on Code: {code}")
    df_t = df.loc[df['ALT_GEO_CODE'] == code]
    df_t = df_t.set_index("CHARACTERISTIC_NAME").T.reset_index()
    df_t["ALT_GEO_CODE"] = df_t.iloc[1,2]
    df_t["GEO_NAME"] = df_t.iloc[0,3]
    df_t.drop([0,1], axis=0, inplace=True)
    df_t.drop("index", axis=1, inplace=True)
    append_df = pd.concat([append_df, df_t])
    
```
